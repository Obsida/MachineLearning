# Описание набора данных "Регрессия Опоссума"

В данном наборе данных представлены различные физические характеристики опоссумов, что позволяет проводить регрессионный анализ. Рассмотрим каждую из переменных:

- **case (идентификатор)**: 
  - Уникальный идентификатор для каждого опоссума в наборе данных. Это целочисленная переменная, которая помогает отслеживать отдельные записи.

- **site (место)**: 
  - Идентификатор места сбора данных о опоссумах. Это целочисленная переменная, которая может указывать на разные географические локации.

- **Pop (популяция)**: 
  - Название популяции, к которой принадлежит опоссум. Это категориальная переменная, которая может быть полезна для анализа различий между популяциями.

- **sex (пол)**: 
  - Пол опоссума, представленный как категориальная переменная ('m' для самцов и 'f' для самок). Эта переменная может быть важной для анализа различий в физических характеристиках между полами.

- **age (возраст)**: 
  - Возраст опоссума в годах. Это непрерывная переменная, которая может использоваться для предсказания других характеристик.

- **hdlngth (длина головы)**: 
  - Длина головы опоссума, измеряемая в миллиметрах. Это целевая переменная для нашего анализа регрессии.

- **skullw (ширина черепа)**: 
  - Ширина черепа опоссума, измеряемая в миллиметрах. Эта переменная может быть связана с длиной головы и другими физическими характеристиками.

- **totlngth (общая длина)**: 
  - Общая длина тела опоссума, измеряемая в миллиметрах. Эта переменная может быть полезна для анализа корреляций с длиной головы.

- **taill (длина хвоста)**: 
  - Длина хвоста опоссума, измеряемая в миллиметрах. Эта характеристика также может быть связана с другими размерами тела.

- **footlgth (длина лапы)**: 
  - Длина лапы опоссума, измеряемая в миллиметрах. Важно учитывать эту переменную при анализе размеров тела.

- **earconch (размер уха)**: 
  - Размер ушной раковины опоссума, измеряемый в миллиметрах. Эта переменная может быть интересна для изучения внешних характеристик.

- **eye (размер глаза)**: 
  - Размер глаза опоссума, измеряемый в миллиметрах. Может быть полезен для анализа визуальных характеристик.

- **chest (обхват груди)**: 
  - Обхват груди опоссума, измеряемый в миллиметрах. Эта переменная может помочь понять общую физическую форму животного.

- **belly (обхват живота)**: 
  - Обхват живота опоссума, измеряемый в миллиметрах. Может быть связан с состоянием здоровья и питанием животного.

# Обучение моделей (models fitting)

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score
# Инициализация логов
log = {}
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['skullw','totlngth','chest','belly','footlgth']], df['hdlngth'], test_size=0.4, random_state=19)
```

## Обучение модели Linear Regression

```python
# Обучение модели Linear Regression
linear = LinearRegression()
linear.fit(X_train, y_train)
linear_predict = linear.predict(X_test)

mse_linear = mean_squared_error(y_test, linear_predict)
mae_linear = mean_absolute_error(y_test, linear_predict)
r2_linear = r2_score(y_test, linear_predict)

# Оценка модели с использованием кросс-валидации для R²
cv_r2_linear = cross_val_score(linear, X_train, y_train, cv=5, scoring='r2').mean()

log['Linear Regression'] = {
    'MAE': mae_linear,
    'MSE': mse_linear,
    'R²': r2_linear,
    'CV Mean R²': cv_r2_linear
}

print(f'\nСредняя абсолютная ошибка (Linear Regression): {mae_linear}')
print(f'Среднеквадратичная ошибка (Linear Regression): {mse_linear}')
print(f'Коэффициент детерминации (Linear Regression): {r2_linear}')
print(f'Среднее значение R² (кросс-валидация): {cv_r2_linear}')
```

## Обучение модели KNeighborsRegressor

```python
# Обучение модели KNeighborsRegressor
knn_regressor = KNeighborsRegressor()
knn_regressor.fit(X_train, y_train)
knn_predict = knn_regressor.predict(X_test)

mse_knn = mean_squared_error(y_test, knn_predict)
mae_knn = mean_absolute_error(y_test, knn_predict)
r2_knn = r2_score(y_test, knn_predict)

# Оценка модели KNN с использованием кросс-валидации для R²
cv_r2_knn = cross_val_score(knn_regressor, X_train, y_train, cv=5, scoring='r2').mean()

log['KNN'] = {
    'MAE': mae_knn,
    'MSE': mse_knn,
    'R²': r2_knn,
    'CV Mean R²': cv_r2_knn
}

print(f'\nСредняя абсолютная ошибка (KNN): {mae_knn}')
print(f'Среднеквадратичная ошибка (KNN): {mse_knn}')
print(f'Коэффициент детерминации (KNN): {r2_knn}')
print(f'Среднее значение R² (кросс-валидация): {cv_r2_knn}')
```

## Обучение модели SVR

```python
# Обучение модели SVR
svr = SVR()
svr.fit(X_train, y_train)
svr_predict = svr.predict(X_test)

mse_svr = mean_squared_error(y_test, svr_predict)
mae_svr = mean_absolute_error(y_test, svr_predict)
r2_svr = r2_score(y_test, svr_predict)

# Оценка модели SVR с использованием кросс-валидации для R²
cv_r2_svr = cross_val_score(svr, X_train, y_train, cv=5, scoring='r2').mean()

log['SVR'] = {
    'MAE': mae_svr,
    'MSE': mse_svr,
    'R²': r2_svr,
    'CV Mean R²': cv_r2_svr
}

print(f'\nСредняя абсолютная ошибка (SVR): {mae_svr}')
print(f'Среднеквадратичная ошибка (SVR): {mse_svr}')
print(f'Коэффициент детерминации (SVR): {r2_svr}')
print(f'Среднее значение R² (кросс-валидация): {cv_r2_svr}')
```

## Обучение модели Random Forest Regressor

```python
# Обучение модели Random Forest Regressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_predict = rf.predict(X_test)

mse_rf = mean_squared_error(y_test, rf_predict)
mae_rf = mean_absolute_error(y_test, rf_predict)
r2_rf = r2_score(y_test, rf_predict)

# Оценка модели Random Forest с использованием кросс-валидации для R²
cv_r2_rf = cross_val_score(rf, X_train, y_train, cv=5, scoring='r2').mean()

log['Random Forest'] = {
    'MAE': mae_rf,
    'MSE': mse_rf,
    'R²': r2_rf,
    'CV Mean R²': cv_r2_rf
}

print(f'\nСредняя абсолютная ошибка (Random Forest): {mae_rf}')
print(f'Среднеквадратичная ошибка (Random Forest): {mse_rf}')
print(f'Коэффициент детерминации (Random Forest): {r2_rf}')
print(f'Среднее значение R² (кросс-валидация): {cv_r2_rf}')
```

```python
# Создание DataFrame из логов
log_df = pd.DataFrame.from_dict(log, orient='index').reset_index()
log_df.rename(columns={'index': 'Model'}, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='R²', y='Model', data=log_df.sort_values(by='R²', ascending=False), color="b")

plt.xlabel('Коэффициент детерминации (R²)')
plt.title('Сравнение коэффициента детерминации моделей')
plt.xlim(0, 1)  
plt.show()
```
